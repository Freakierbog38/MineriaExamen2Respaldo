```python
import pandas as pd
import numpy as np

df = pd.read_csv('weatherAUS.csv')
df.head()
```

```python
df.info()
```

```python
to_drop = [
    'Date',
    'Location',
    'WindGustDir',
    'WindDir9am',
    'WindDir3pm',
    'RainToday',
    'RISK_MM'
]

df.drop(to_drop, inplace=True, axis=1)
```

```python
df.head()
```

```python
rs = np.random.RandomState(0)
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
cleaned = df.fillna(0)
cleaned = cleaned.replace({'RainTomorrow' : {'No':-1, 'Yes':1}})
```

```python
cleaned
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X, y = cleaned.iloc[:,:-1].values, cleaned.iloc[:,-1].values

train, test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)

# standardize the features
sc = StandardScaler()
sc.fit(train)
X_train_std = sc.transform(train)
X_test_std = sc.transform(test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=5, random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
all_accuracies = cross_val_score(estimator=classifier, X=X_train_std, y=y_train, cv=5)
```

```python
print(all_accuracies.mean())
```

```python
from plot_regions import plot_decision_regions
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
lr = LogisticRegression()

X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
lr.fit(X_train_pca, y_train)
plot_decision_regions(X_train_pca, y_train, classifier=lr)

plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(loc='lower left')
plt.show()
```

```python
# test the transformed test dataset
plot_decision_regions(X_test_pca, y_test, classifier=lr)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(loc='lower left')
plt.show()
```

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV

grid_params = {
    'n_estimators': [1, 3, 5, 8, 10],
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False]
}

gd_sr = GridSearchCV(estimator=classifier,
                    param_grid=grid_params,
                    scoring='accuracy',
                    cv=5,
                    n_jobs=1)
```

```python
gd_sr.fit(X_train_std, y_train)
```

```python
best_parameters= gd_sr.best_params_
print(best_parameters)
```
